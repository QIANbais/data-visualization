### Import csv into Pandas DataFrame

```python
df = pd.read_csv("04-01-2020.csv")
df['Confirmed']=df.Confirmed.astype(float)
df.head()
```

### Creating the Folium interactive map

```python
world_map = folium.Map(location=[11,0], tiles="cartodbpositron", zoom_start=2, max_zoom = 6, min_zoom = 2)
# iterate over all the rows of confirmed_df to get the lat/long
for i in range(0,len(df)):
    folium.Circle(
        location=[df.iloc[i]['Lat'], df.iloc[i]['Long_']],
        fill=True,
        radius=(int((np.log(df.iloc[i,7]+1.00001)))+0.2)*30000,
        color='red',
        fill_color='red',
    ).add_to(world_map)
    
world_map
```

#### the visualization above display the confirmed cases across world on 04/01/202
